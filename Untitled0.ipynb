Voy a arrancar a trabajar con lo que sería el primer bloque del ETL: Extracción.

Lo que voy a hacer es escribr código base para traer los datos desde la API (es decir, voy a ejecutar un Script) convertirlos a un DataFrame y guardarlos en la carpeta data/raw/.

Con este código se va a cumplir lo siguiente:

* Importar datos desde la API.

* Convertir a DataFrame.

* Guardar el archivo original para tener un respaldo.

```python
import pandas as pd
import requests

# URL directa al archivo JSON en GitHub (en modo raw)
raw_url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Descargar los datos
response = requests.get(raw_url)

if response.status_code == 200:
    data = response.json()
    print("Datos descargados desde GitHub OK")

    # Convertir a DataFrame
    df = pd.DataFrame(data)
    print(f"DataFrame con {df.shape[0]} filas y {df.shape[1]} columnas")

    # Guardar el archivo original
    df.to_csv("telecomx_raw.csv", index=False)
    print("Archivo guardado como telecomx_raw.csv")

else:
    raise Exception(f"Error al descargar datos: {response.status_code}")
```

Extracción lista, los datos están en un DataFrame y respaldados en CSV.

### Observacion

Acá hay algo clave: las columnas customer, phone, internet y account no son valores simples, sino diccionarios dentro de celdas (objetos JSON anidados).

Esto significa que primero tenemos que "aplanar" estas columnas para que cada campo sea su propia columna en el DataFrame antes de analizar nada.

### Con este paso:

Vamos a tener todas las variables demográficas, de teléfono, internet y cuenta separadas.

La columna Churn seguirá como indicador de si el cliente se fue o no.

El DataFrame pasará de 6 columnas a unas 15–20 (dependiendo de los campos originales).

```python
import pandas as pd

# Aplanar columnas JSON anidadas
customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])

# Renombrar columnas para evitar conflictos
customer_df.columns = [f"customer_{col}" for col in customer_df.columns]
phone_df.columns = [f"phone_{col}" for col in phone_df.columns]
internet_df.columns = [f"internet_{col}" for col in internet_df.columns]
account_df.columns = [f"account_{col}" for col in account_df.columns]

# Unir todo en un solo DataFrame
df_clean = pd.concat(
    [df[['customerID', 'Churn']], customer_df, phone_df, internet_df, account_df],
    axis=1
)

# Guardar el resultado
df_clean.to_csv("telecomx_flatten.csv", index=False)
print("Datos aplanados guardados como telecomx_flatten.csv")
print(f"Nuevo DataFrame con {df_clean.shape[0]} filas y {df_clean.shape[1]} columnas")
```

### 2. Exploración y limpieza incial adaptada al nuevo DataFrame

```python
# =======================
# 1. Exploración inicial
# =======================
print("Primeras filas del dataset aplanado:")
display(df_clean.head())

print("\nInformación general:")
df_clean.info()

print("\nValores nulos por columna:")
print(df_clean.isnull().sum())

print("\nResumen estadístico:")
display(df_clean.describe(include='all'))

# =======================
# 2. Limpieza de datos
# =======================

# 2.1 Eliminar duplicados
df_clean.drop_duplicates(inplace=True)

# 2.2 Convertir Yes/No a valores binarios (0/1)
yes_no_cols = df_clean.select_dtypes(include='object').columns
for col in yes_no_cols:
    if df_clean[col].isin(['Yes', 'No']).all():
        df_clean[col] = df_clean[col].map({'Yes': 1, 'No': 0})

# 2.3 Asegurar tipos numéricos donde corresponda
# (pandas intentará convertir las que pueda)
for col in df_clean.columns:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='ignore')

# 2.4 Limpiar nombres de columnas
df_clean.columns = df_clean.columns.str.strip().str.lower().str.replace(" ", "_")

print("\nDespués de limpieza:")
df_clean.info()

# Guardar dataset limpio
df_clean.to_csv("telecomx_clean.csv", index=False)
print("\nDatos limpios guardados como telecomx_clean.csv")
```

### Con este bloque se logró:

* Tener todo sin duplicados.

* Variables categóricas Yes/No convertidas a binario (más fácil para análisis y modelos).

## Nuevo análisis - [ESTANDARIZACION Y TRANSFORMACION DE DATOS (OPCIONAL)]

====================================

La idea de esta tarjeta es limpiar y transformar datos que son textuales para que después sean fáciles de procesar (por ejemplo: cambiar "Yes"/"No" → 1/0, traducir nombres de columnas al español, etc.).

* Propuesta de trabajo:

a) Convertir valores binarios a 0 y 1

* Columnas como "PaperlessBilling", "Churn", "Partner", "Dependents", "PhoneService", etc.

* Ejemplo: "Yes" → 1, "No" → 0.

b) Estandarizar valores categóricos simples

* Por ejemplo "Male" → "Hombre", "Female" → "Mujer".

* "Month-to-month" → "Mensual", "One year" → "Anual", "Two year" → "Bianual".

c) Opcional: renombrar columnas al español

* MonthlyCharges → CargosMensuales

* TotalCharges → CargosTotales

* Contract → Contrato

```python
import ast  # para convertir strings que parecen diccionarios a diccionarios reales

# Función para parsear strings a diccionarios si es necesario
def parse_dict(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return {}
    return x

# Aplicamos parseo
df_clean['customer_dict'] = df_clean['customer'].apply(parse_dict)
df_clean['account_dict'] = df_clean['account'].apply(parse_dict)
df_clean['phone_dict'] = df_clean['phone'].apply(parse_dict)
df_clean['internet_dict'] = df_clean['internet'].apply(parse_dict)

# Extraemos columnas planas
df_clean['gender'] = df_clean['customer_dict'].apply(lambda x: x.get('gender', '').capitalize())
df_clean['Contract'] = df_clean['account_dict'].apply(lambda x: x.get('Contract', ''))
df_clean['PaymentMethod'] = df_clean['account_dict'].apply(lambda x: x.get('PaymentMethod', ''))
df_clean['PhoneService'] = df_clean['phone_dict'].apply(lambda x: x.get('PhoneService', ''))
df_clean['MultipleLines'] = df_clean['phone_dict'].apply(lambda x: x.get('MultipleLines', ''))
df_clean['InternetService'] = df_clean['internet_dict'].apply(lambda x: x.get('InternetService', ''))

# Ahora hacemos el recuento de evasión por variable categórica
categorical_vars = ['gender', 'Contract', 'PaymentMethod', 'PhoneService', 'MultipleLines', 'InternetService']

for var in categorical_vars:
    print(f"\nRecuento de evasión según '{var}':")
    display(df_clean.groupby([var, 'Churn']).size().unstack(fill_value=0))

    # Gráfico de barras
    plt.figure(figsize=(10,4))
    sns.countplot(data=df_clean, x=var, hue='Churn', palette='Set2')
    plt.title(f"Evasión por {var}")
    plt.xticks(rotation=45)
    plt.show()
```
